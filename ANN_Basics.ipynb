```python
import pandas as pd
```

```python
data = pd.read_csv("/content/Churn_Modelling.csv")
data.head()
```

```python
data = data.drop(["RowNumber","CustomerId","Surname"],axis=1)
```

```python
x = data.drop(["Exited"], axis=1)
y = data["Exited"]
```

```python
geo = pd.get_dummies(x["Geography"])
gen = pd.get_dummies(x["Gender"])
```

```python
geo
```

```python
x = pd.concat([x,geo,gen], axis=1)
```

```python
x = x.drop(["Geography","Gender"], axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, LeakyReLU, PReLU, ELU, Dropout
```

Applying Model

```python
model = Sequential()
```

Adding input and 1st hidden layer

```python
model.add(Dense(units=6, kernel_initializer='he_uniform', activation="relu", input_dim=13))
```

Adding 2nd hidden layer

```python
model.add(Dense(units=6, kernel_initializer='he_uniform', activation="relu"))
```

Adding output layer

```python
model.add(Dense(units=1, kernel_initializer='glorot_uniform', activation="sigmoid"))
```

compiling our ann model

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

Fitting our model with training data

```python
# model_history = model.fit(x_train, y_train, validation_split= 0.33, batch_size=10, nb_epoch=100)
model_history = model.fit(x_train, y_train,validation_split=0.33, batch_size = 10, epochs = 100)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
```
